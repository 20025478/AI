```python
import requests
from bs4 import BeautifulSoup
import time
#import schedule
def scrape_weather():

  url = 'https://weather.com/en-IE/weather/today/l/EIXX0014:1:EI?Goto=Redirected'
  html = requests.get(url)
  soup = BeautifulSoup(html.text, 'html.parser')
  location_tag =soup.find('h1', class_ ='CurrentConditions--location--1YWj_')
  timestamp_tag =soup.find('span', class_='CurrentConditions--timestamp--1ybTk')
  temperature_tag = soup.find('span', class_='CurrentConditions--tempValue--MHmYY')
  day_night_temp_tag =soup.find('div', class_ ='CurrentConditions--tempHiLoValue--3T1DG')

  if location_tag and timestamp_tag and temperature_tag and day_night_temp_tag:
     location = location_tag.text.strip()
     timestamp = timestamp_tag.text.strip()
     temperature= temperature_tag.text.strip()
     day_night_temp =day_night_temp_tag.text.strip()
     print(f"{location} {timestamp}\n{temperature}\n{day_night_temp}")
     print('-' * 30)
  else:
     print("information not found")

  alert_headline_tag = soup.find('h2',class_='AlertHeadline--alertText--38xov')
  if alert_headline_tag:
     alert_headline = alert_headline_tag.text.strip()
     print(f"\033[1mAlert Headline: {alert_headline}\033[0m")
  else:
     print(f"\033[1mNo alert found\033[0m")

  climate_info_tag = soup.find("div", class_='Card--content--1GQMr PrecipIntensityCard--PrecipIntensityCardContent--2VnJ1')
  if climate_info_tag:
     climate_info = climate_info_tag.text.strip()
     print(f"\033[1mClimate info: {climate_info}\033[0m")

if 'weather_data_feched' not in locals():
    scrape_weather()
    weather_data_feched = True

#schedule.every(10).minutes.do(scrape_weather)
while True:
    #schedule.run_pending()
    time.sleep(60)
    scrape_weather()

```
